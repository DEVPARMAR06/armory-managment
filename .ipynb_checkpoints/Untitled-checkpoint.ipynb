```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Mar 21 18:55:32 2021

@author: devumeshkumarparmar
"""


#we want to user place because we want a back ground image

from tkinter import *
from PIL import Image,ImageTk
import tkinter.messagebox as tkmb
from tkinter.ttk import *

# defs
def userr(event):
    user1.configure(state=NORMAL)
    user1.delete(0,END)
    usercheck=True

def passw(event):
    pas1.configure(state=NORMAL)
    pas1.delete(0, END)
    passcheck=True

def admin():
    u=user.get()
    p=pas.get()
    f=open('Admin_Data.txt','r')
    b=0
    for line in f.readlines():
        a=line.split(' ')
        for i in range(1):
            if a[0]==u and a[1]==p:
                tkmb.showinfo(title='Message Box', message='Varified User', icon='info')
                b=b+1
    if b==0:
        tkmb.showinfo(title='Message Box', message='INCORRECT DETAILS', icon='warning')   
            
            
    
    


#create instance
root=Tk()

#title text
root.title("Admin Login")

#geomatry width x length
#set max and min width
root.geometry("1100x688")
root.minsize(1100,688)
root.maxsize(1100,688)

#back ground image For Jpg Use ImageTk
photo=Image.open("192866.jpg")
bg=ImageTk.PhotoImage(photo)

#create a lable size of a window
#we use lable as a back ground image

d1=Label(image=bg)
d1.pack()
#root.attributes('-alpha',0.5)  #to use whole screen transparant
#-----------------------------------
# Create A Label in grid

#for palceholder
usercheck=False
passcheck=False

#No Need To Use Label 

#username=Label(root,text="User_Name:")
#username.place(x=300, y=150)

#password=Label(root,text="User_Password")
#password.place(x=500, y=150)

#create a entry width

user=StringVar()
user1=Entry(root,textvariable=user)
user1.place(x=300, y=250)
# for place holder
user1.insert(0, "UserName")
user1.configure(state=DISABLED)
user1.bind("<Button>",userr)
#------
pas=StringVar()
pas1=Entry(root,textvariable=pas)
pas1.place(x=300, y=300)
# for place holder-2
pas1.insert(0, "PassWord")
pas1.configure(state=DISABLED)
pas1.bind("<Button>",passw)

# For Submit Button
# For submit image
#style = Style()
#style.configure('W.TButton', font =
               #('calibri', 5, 'bold', 'underline'),bg="black",borderwidth=0,foreground="red",relief="FLAT")
#style.map('W.TButton', foreground = [('active', '! disabled', 'green')],
                     #background = [('active', 'black')])
sub=PhotoImage(file="login-button-png-18035.png")

b1=Button(root,image=sub,command=admin)
b1.place(x=300,y=350)
root.mainloop()
```
